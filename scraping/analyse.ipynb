```python
t_globalT_dict = {}
for i, t in enumerate(to_keep):
    n = int(input(f'{i}/{len(to_keep)}: {t} --> '))
    if n == 0:
        t_globalT_dict[t] = 'Computer Vision'
    elif n == 1:
        t_globalT_dict[t] = 'Contrastive learning'
    elif n == 2:
        t_globalT_dict[t] = 'NLP'
    elif n == 3:
        t_globalT_dict[t] = 'Reinforcement learning'
```

```python
filtered_data = {}
for u, (a, t) in tqdm(data.items()):

    new_t = set()
    for tt in t:
        if tt in t_globalT_dict:
            new_t.add(t_globalT_dict[tt])
    
    new_t = list(new_t)

    if len(new_t) == 2:
        if 'NLP' in new_t and 'Computer Vision' in new_t:
            if new_t.count('NLP') >=  new_t.count('Computer Vision'):
                new_t = ['NLP']
            else:
                new_t = ['Computer Vision']
        
        elif 'Contrastive learning' in new_t:
            new_t = ['Contrastive learning']

        elif 'Reinforcement learning' in new_t:
            new_t = ['Reinforcement learning']

    elif len(new_t) == 3:
        if 'Contrastive learning' in new_t:
            new_t = ['Contrastive learning']
        elif 'Reinforcement learning' in new_t:
            new_t = ['Reinforcement learning']
    
    if new_t:
        filtered_data[u] = [a, new_t[0]]
```
